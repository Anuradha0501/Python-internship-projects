```python
import sys
from tkinter import *
from collections import OrderedDict

def getnote(tone):
    noted = ['E ', 'F ', 'F#', 'G ', 'Ab','A ', 'Bb', 'B ', 'C ', 'Db', 'D ', 'Eb']
    return noted[tone % 12]

def getoffset_tone(tone):
    scaleref = {'E ': 0, 'F ': 1, 'F#': 2, 'G ': 3, 'Ab': 4, 'A ': 5, 'Bb': 6, 'B ': 7, 'C ': 8, 'Db': 9, 'D ': 10,'Eb': 11}
    return scaleref[tone]

scales = OrderedDict([
    ('Major', [0, 2, 2, 1, 2, 2, 2, 1]),
    ('Natural minor', [0, 2, 1, 2, 2, 1, 2, 2]),
    ('Harmonic minor', [0, 2, 1, 2, 2, 1, 3, 1]),
    ('Melodic minor', [0, 2, 1, 2, 2, 2, 2, 2]),
    ('Dorian mode', [0, 2, 1, 2, 2, 2, 1, 2]),
    ('Phrygian mode', [0, 1, 2, 2, 2, 1, 2, 2]),
    ('Lydian mode', [0, 2, 2, 2, 1, 2, 2, 1]),
    ('Locrian mode', [0, 1, 2, 2, 1, 2, 2, 2]),
    ('Minor pentatonic', [0, 3, 2, 2, 3, 2]),
    ('Pentatonic', [0, 2, 2, 3, 2, 3]),
    ('Blues', [0, 3, 2, 1, 1, 3]),
    ('Major chord', [0, 4, 3]),
    ('Minor chord', [0, 3, 4]),
    ('Diminished chord', [0, 3, 3]),
    ('Augmented chord', [0, 4, 4]),
    ('Sus2 chord', [0, 2, 5]),
    ('Sus4 chord', [0, 5, 2]),
    ('Maj7 chord', [0, 4, 3, 4]),
    ('min7 chord', [0, 3, 4, 3]),
    ('7 chord', [0, 4, 3, 3]),
    ('Maj9 chord', [0, 4, 3, 4, 3]),
    ('m9 chord', [0, 3, 4, 3, 4]),
    ('11 chord', [0, 4, 3, 3, 4, 3]),
    ('Maj11 chord', [0, 4, 3, 4, 3, 3]),
    ('min11 chord', [0, 3, 4, 3, 4, 3]),
])

def makescale(keyroot, keyopt):
    keywheel = []
    keywheel.extend(scales[keyopt])
    filler = 0
    scale = []
    lenvar = len(keywheel) 
    for inte in range(lenvar):
        filler += keywheel[inte % len(keywheel)]
        scale.append(int(filler + getoffset_tone(keyroot)))
    return scale

scale = makescale('E ', 'Major chord')

e = 0

offsetArray = [e, e + 7, e + 3, e + 10, e + 5, e]

wind = Tk()

variable = StringVar(wind)
variable.set('E ')
variable2 = StringVar(wind)
variable2.set('Major chord')

def resettable():
    for i in range(0, 25):
        for gss in range(0, 6):
            Label(wind, text=getnote(i + offsetArray[gss])).grid(row=gss + 2, column=i + 1, padx=0, pady=0)

def applyit(val):
    tonic = str(variable.get())
    key = str(variable2.get())
    scale = makescale(tonic,key)
    ournotes = []

    for notes in scale:
        ournotes.append(getnote(notes))
   
    for i in range(0, 25):
        for gss in range(0, 6):
            start = offsetArray[gss]
            
            if tonic == getnote(i + start % 12):
                Label(wind, text=getnote(i + start), bg='grey').grid(row=gss + 2, column=i + 1,padx=0, pady=0)

            elif getnote(i + start) in ournotes:
                Label(wind, text=getnote(i + start), bg='pink').grid(row=gss + 2, column=i + 1,padx=0, pady=0)

            else:
                Label(wind, text=getnote(i + start)).grid(row=gss + 2, column=i + 1, padx=0, pady=0)


if __name__ == "__main__":
    wind.geometry('630x300')
    wind.title('Scale Chart for stringed instruments ')
    ourx = 40
    oury = 20

    for i in range(0, 25):
        Label(wind, text=i, font='Calibri').grid(row=0, column=i + 1, padx=0, pady=10)
        Label(wind, text=i, font='Calibri').grid(row=9, column=i + 1, padx=0, pady=10)

    stringarray = ['E', 'B', 'G', 'D', 'A', 'E']
    for gss in range(0, 6):
        Label(wind, text=stringarray[gss], font="Calibri").grid(row=gss + 2, column=0, padx=10, pady=0)

    applyit("")

    keymenu = OptionMenu(wind, variable, 'E ', 'F ', 'F#', 'G ', 'Ab', 'A ', 'Bb', 'B ', 'C ', 'Db', 'D ','Eb', command=applyit).place(x=ourx * 2, y=oury * 12)
    scalemenu = OptionMenu(wind, variable2, *scales.keys(), command=applyit).place(x=ourx * 4, y=oury * 12)
    resetbutton = Button(wind, text=' Reset ', command=resettable).place(x=ourx * 10, y=oury * 12)
    
    wind.mainloop()
```
