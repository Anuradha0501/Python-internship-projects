```python
import random 
import tkinter
from tkinter import *
from tkinter import ttk

#window attributes
wind=Tk()
wind.geometry("800x500")
wind.title("Python Story Generator!")
wind.configure(bg='Pink')


#Title and information
til=Label(wind,justify=CENTER,text="Python Story Generator!",background="Black",fg="White",width=50) 
til.place(x=-60,y=2)
til.config(font=("Courier", 15,"bold"))
ins=Label(wind,text="Hello Reader!!! ")
ins.place(x=0,y=30)
ins1=Label(wind,justify=CENTER,text="Here we create an our story.",
           background="Black",fg="White",width=50) 
ins1.place(x=0,y=50)


info=Label(wind,text="Now let me know you are girl or boy? : ",background = "Black",fg="light blue") 
info.place(x=20,y=100) 
info.config(font=("Courier", 11,"bold"))
entry = Entry(wind).place(x=400,y=100)

name=Label(wind,text="What is your name : ",background = "Black",fg="light blue") 
name.place(x=20,y=130) 
name.config(font=("Courier", 11,"bold"))
name1=Entry(wind) 
name1.place(x=400,y=130)
frnd=Label(wind,text="What is your friend name? : ",background = "Black",fg="light blue") 
frnd.place(x=20,y=160) 
frnd.config(font=("Courier", 11,"bold"))
frnd1=Entry(wind) 
frnd1.place(x=400,y=160)
hobbies=Label(wind,text="What is your hobby e.g.-reading book? : ",background = "Black",fg="light blue") 
hobbies.place(x=20,y=190) 
hobbies.config(font=("Courier", 11,"bold"))
hobbies1=Entry(wind) 
hobbies1.place(x=400,y=190)



#Generate story
def gen_story1():
    wind1=Tk()
    wind1.geometry("600x400")
    wind1.title("Python Story Generator!")
    wind1.configure(bg='White')
    txt = Text(wind1)
    
    pos = ["nice", "amazing","great","good", "interesting"]
    neg = ["careless", "dishonest","useless"]
    word1 = ["comment","question"]
    word2 = ["birds", "animals", "programming", "finding new opportunity", "problem solving ability", "new technology"]
    word3 = ["liked", "disliked"]
    if entry=="girl":
        Story = ["Yesterday while ",'hobbies1',"",'name1',"finds that her friend",'frnd1',
                 "send her some articles related to", 'word2',". She finds it"
                 ,'pos',"so she decided to learn more about it",
                 ". She called her friend to send more articles. But as she was so"
                 ,'neg',"so se don't send her article. After that",'name1'," ",'word1',"her friend on her laziness."]
    else:  
        Story = ["Yesterday while browsing about",'hobbies1',"",'name1',"noticed that his  friend",'frnd1',"posted a new",
                 'word1',"about",'word2', "He",'word3',"it. His friend challenged him a quiz about that post.",'name1',
                 "finds it",'pos', "so he agreed to take quiz and after sometime his friend noticed that he",'word1',
                 "a",'neg',"quiz as",'name1',"already knows more than him"]

    for item in Story:
        if item == 'name1' : Story[Story.index(item)] = name1.get()
        elif item == 'hobbies1' : Story[Story.index(item)] = hobbies1.get()
        elif item == 'frnd1' : Story[Story.index(item)] = frnd1.get()
        elif item == 'pos' : Story[Story.index(item)] = random.choice(pos)
        elif item == 'neg' : Story[Story.index(item)] = random.choice(neg)
        elif item == 'word1' : Story[Story.index(item)] = random.choice(word1)
        elif item == 'word2' : Story[Story.index(item)] = random.choice(word2)
        elif item == 'word3' : Story[Story.index(item)] = random.choice(word3)
        else: continue
   
    #Story
    #a = ' \n'.join(Story)
    #print(e.get("".join(Story)))
    txt.insert(INSERT, " ".join(item for item in Story))
    txt.pack()
      
        

btn=Button(wind, text="Create Story!",activebackground="blue",background = "white",command=gen_story1) 
btn.place(x=200,y=230)

#This will show window
wind.mainloop()
```
